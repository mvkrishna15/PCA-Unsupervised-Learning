```python
from sklearn.pipeline import Pipeline
from sklearn .preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn. model_selection import train_test_split
```

```python
##Standardization : scaling down the data in a way where mean=0, std dev=1
```

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import *
```

```python
pca_pipe=Pipeline([('sr',StandardScaler()),
                  ('pca',PCA(n_components=25))])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=23)
```

```python
x_pca=pca_pipe.fit_transform(x_train)
xt_pca=pca_pipe.transform(x_test)
```

```python
dt=DecisionTreeClassifier(random_state=23)
dt.fit(x_pca,y_train)
```

```python
y_pred=dt.predict(xt_pca)
```

```python
accuracy_score(y_pred,y_test)*100
```
